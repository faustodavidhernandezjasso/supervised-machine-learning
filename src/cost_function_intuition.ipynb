```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm 
from matplotlib.ticker import LinearLocator
import numpy as np

# Load data set
df = pd.read_csv('data_set/first_data_set/test.csv')
df.plot.scatter(x='x', y='y')
```

```python
x = np.array(df['x'])
y = np.array(df['y'])
m = x.shape[0]
print(f"Number of instances is: {m}")
```

```python
# Plot the data points
plt.scatter(x, y, marker='x', c='r')
# Set the title
plt.title("Dummy test")
# Set the y-axis label
plt.ylabel("Label of y")
# Set the x-axis label
plt.xlabel("Label the x-axis")
```

```python
def model(w, b):
    return w * x + b

def simplified_model(w):
    return w * x

def cost_function(model):
    cost = 0
    for i in range(m):
        error = (model[i] - y[i]) ** 2
        cost += error
    return (1 / m) * cost
```

```python
# Our first parameter
w_1 = 0.8
# Our first prediction
prediction_w_1 = simplified_model(w_1)
# Plot our model prediction
plt.plot(x, prediction_w_1, c='b', label='Our first prediction')
# Plot the data points
plt.scatter(x, y, marker='x', c='r', label='Actual value')
# Set the title
plt.title("Dummy test")
# Set the y-axis label
plt.ylabel("y")
# Set the x-axis label
plt.xlabel("x")
plt.legend()
plt.show()
```

```python
# Compute the J(w) function
cost_function_w_1 = cost_function(prediction_w_1)
print(f"The cost of our prediction is: {cost_function_w_1}")
plt.scatter(w_1, cost_function_w_1, marker='d', c='pink')
# Set the y-axis label
plt.ylabel("J(w)")
# Set the x-axis label
plt.xlabel("w")
```
